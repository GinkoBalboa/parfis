# Scattering after collision

In the collision simulation, once the type of collision is determined, it is then necessary to 
determine the direction of the projectile after scattering. We derive the scattering angle in the 
general case where the masses and velocities of projectile and target particles are comparable. Later
we introduce the approximation for lightweight projectiles with orders of magnitude greater velocity
than the targets (electron-neutral collisions). We make our calculations in the center of mass
CM reference frame, where the static center of mass is at the origin of the coordinate system.

## Center of mass reference frame

The CM frame is defined as the inertial frame in which the sum of the linear momenta of all particles 
is equal to 0. We denote the laboratory reference system with L, and with C the CM reference frame. 
The velocity of the mass center is:

$$
V_C = \frac{m_1 v_{L1} + m_2 v_{L2}}{m_1 + m_2},
$$

where with number 1 is denoted the projectile particle and with 2 the target particle. In the CM frame
the velocities are:

$$
\begin{align}
v_{C1} = v_{L1} - V_C \\
v_{C2} = v_{L2} - V_C
\end{align}
$$

which, for the momentum in CM implies:

$$
p_{C1} = -p_{C2}
$$

### Special case: electron impacting gas molecule

Because the mass of the electron is very much smaller than that of the molecule, and the velocity 
of the electron is much bigger than that of the molecule we can consider that the molecule is at 
rest in the LAB frame. This simplifies things a lot since the scattering is reduced to selecting the
scattering angle $\theta$ and the azimuthal angle $\phi$.

For anisotropic scattering (we assume only azimuthal isotropy), the angle $\theta$ is not uniformly 
distributed but an angular dependence $I(\theta)$ exists. A beam of incident scatterers with a 
given flux (number of particles per unit area per unit time) impinges on the target $I_{inc}$. 
The number of particles per unit time which are detected in a small ring region of the angle $\theta$
can be counted as:

$$
\frac{I(\theta)}{I_{inc}} = \frac{I(\theta)}{\int_0^{\pi} I(\theta)sin(\theta)d\theta\ }
$$
