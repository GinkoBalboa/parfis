# Scattering after collision

In the collision simulation, once the type of collision is determined, it is then necessary to 
determine the direction of the projectile after scattering.

## Differential cross-section

$$
\renewcommand{\d}{\mathrm{d}}
$$

When a cross-section is specified as the differential limit of a function of some final-state variable, 
such as particle angle or energy, it is called a differential cross-section. When a differential cross-section 
is integrated over all scattering angles we get the integral cross-section (when talking of cross-sections we 
refer to the integral cross-sections).

Consider a single particle, scattered off a single stationary target. A spherical coordinate system is used, 
with the target placed at the origin and the z-axis of this coordinate system aligned with the incident beam. 
The angle $\theta$ is the scattering angle, measured between the incident beam and the scattered beam, 
and the angle $\phi$ is the azimuthal angle. An incident particle scatters if it goes through the collision 
cross-section, but the angle at which it deflects depends on the position where the interaction happens. The 
differential angular range of the scattered particle at angle $\theta$ is the solid angle element 
$\d\Omega = \sin(\theta)\d\theta\d\phi$. The differential cross section is the quotient $\frac{\d\sigma}{\d\Omega}$.

Often we have a cylindrical symmetric situation where the scattering angle depends only on the impact parameter $b$ 
(distance from the z-axis as shown in figure 1) and not on the azimuth.

### Special case: electron impacting gas molecule

Because the mass of the electron is very much smaller than that of the molecule, and the velocity 
of the electron is much bigger than that of the molecule we can consider that the molecule is at 
rest in the LAB frame. This simplifies things a lot since the scattering is reduced to selecting the
scattering angle $\theta$ and the azimuthal angle $\phi$.

For anisotropic scattering (we assume only azimuthal isotropy), the angle $\theta$ is not uniformly 
distributed but an angular dependence $I(\theta)$ exists. A beam of incident scatterers with a 
given flux $I_{inc}$ (number of particles per unit area per unit time) impinges on the target. 
The number of particles per unit time detected in the area of thin ring region, at the angle $\theta$,
can be counted as:

$$
\frac{I(\theta)}{I_{inc}} = \frac{I(\theta)}{\int\limits_0^{\pi} I(\theta)\sin(\theta)d\theta\ }
$$

since the ring of thickness $d\theta$ at the angle $\theta$ has a radius of $r=R\sin(\theta)$. 
Given this radius, the area of the ring is $2\pi r d\theta$.
